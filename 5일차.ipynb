### 4. 고객별 여행보험 가입 여부 예측 모형을 만들고, 가장 높은 Accuracy 값을 가지는 최종 모델을 도출하시오.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train = pd.read_csv('data/304_x_train.csv')
x_test = pd.read_csv('data/304_x_test.csv')
y_train = pd.read_csv('data/304_y_train.csv')

# print(x_train.info())
# print(x_train.describe())

# print(x_train.isnull().sum())
# print(x_test.isnull().sum())
# print(y_train.isnull().sum())

col_del = ['ID']
col_num = ['Age','AnnualIncome','FamilyMembers','ChronicDiseases']
col_cat = ['Employment Type','GraduateOrNot','FrequentFlyer','EverTravelledAbroad']
col_tag = ['TravelInsurance']

x = pd.concat([x_train, x_test])

for col in col_cat:
    le = LabelEncoder()
    le.fit(x[col])
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])

x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train[col_tag], test_size = 0.3, stratify = y_train[col_tag])

scaler = StandardScaler()
x_tr[col_num] = scaler.fit_transform(x_tr[col_num])
x_val[col_num] = scaler.fit_transform(x_val[col_num])
x_test[col_num] = scaler.fit_transform(x_test[col_num])

model_rf = RandomForestClassifier()
model_rf.fit(x_tr, y_tr)

y_pred_rf = model_rf.predict_proba(x_val)
print(roc_auc_score(y_val, y_pred_rf[:,1]))

pred = model_rf.predict(x_test)
pd.DataFrame({'y_pred':pred}).to_csv('submit/004000006.csv')
```

### 2. 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련데이터로 추출 후, 전체 방 개수 컬럼을 의미하는 변수의 결측치를 변수의 중앙값으로 대체한 데이터 세트를 구성한다. 결측치 대체 전의 전체 방 개수 변수 표준편차 값과 대체 후의 표준편차 값의 차이에 대한 절댓값을 계산하시오.

```python
import pandas as pd
import numpy as np

room = pd.read_csv('data/202_housing.csv')
room = room.loc[:len(room)*0.8,]

before = np.std(room['total_bedrooms'])
before

room_remove = room
median = room_remove['total_bedrooms'].median()
room_remove['total_bedrooms'] = room_remove['total_bedrooms'].fillna(median)
after = np.std(room_remove['total_bedrooms'])

print(abs(after-before))
```

### 3. 2번 문항의 데이터에서 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오.(이상값은 평균에서 1.5 표준편차를 초과하거나 미만인 값의 범위로 정함)

```python
pop = pd.read_csv('data/203_housing.csv')

mean = pop['population'].mean()
std = pop['population'].std()

upper = mean+1.5*std
lower = mean-1.5*std

pop[(pop['population']>upper)|(pop['population']<lower)]['population'].sum()
```

### 4. 주어진 훈련 데이터 세트를 사용하여 고객이 주문한 물품의 정시 도착 여부를 예측하고 해당 예측 결과를 csv 파일로 제출하시오.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

x_train = pd.read_csv('data/204_x_train.csv')
x_test = pd.read_csv('data/204_x_test.csv')
y_train = pd.read_csv('data/204_y_train.csv')

#print(x_train.info())
#print(x_train.describe())
#print(y_train.info())

#print(x_train.isnull().sum())
#print(x_test.isnull().sum())
#print(y_train.isnull().sum())

col_del = ['ID']
col_num = ['Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Discount_offered','Weight_in_gms']
col_cat = ['Warehouse_block','Mode_of_Shipment','Product_importance','Gender']
col_tag = ['Reached.on.Time_Y.N']

x = pd.concat([x_train,x_test])

for col in col_cat:
    le = LabelEncoder()
    le.fit(x[col])
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])

x_train = x_train.drop(col_del, axis = 1)
x_test = x_test.drop(col_del, axis = 1)

x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train[col_tag], test_size = 0.3, stratify = y_train[col_tag])

scaler = StandardScaler()
scaler.fit(x_tr[col_num])
x_tr[col_num] = scaler.transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test[col_num] = scaler.transform(x_test[col_num])

model_rf = RandomForestClassifier()
model_rf.fit(x_tr, y_tr)

y_pred_rf = model_rf.predict_proba(x_val)
print(roc_auc_score(y_val, y_pred_rf[:,1]))

pred = model_rf.predict_proba(x_test)
pd.DataFrame({'y_pred':pred[:,1]}).to_csv('submit/004000007.csv')
```
